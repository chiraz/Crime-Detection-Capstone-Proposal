```python
import pandas as pd
```

```python
from scrapy import Selector
```

```python
import requests
```

****

```python
BASE_URL = 'https://www.assarih.com/'
URL2 = 'https://www.assarih.com/r4-%D9%85%D8%AC%D8%AA%D9%85%D8%B9'
URL2 = 'https://www.assarih.com/r4-مجتمع'
URL2_xpath_title = "//div[@class='article']//div[@class='article-title']/a/text()"
URL2_xpath_link = "//div[@class='article']//div[@class='article-title']/a/@href"
URL2_xpath_date = "//div[@class='article']//div[@class='article-title']/p/text()"
URL2_xpath_content = "//div[@class='last-article']//text()"
```

```python
## harvest list of articles' urls from main society pages of the Assarih newspaper

res = []
for i in range(1,100):
    url = URL2+'?page={}'.format(i)
    html = requests.get(url).content
    sel = Selector(text = html)
    u = sel.xpath(URL2_xpath_title).extract()
    v = sel.xpath(URL2_xpath_link).extract()
    w = sel.xpath(URL2_xpath_date).extract()
    z = [y for y in [x.strip() for x in w] if y!='']
    if len(u)!=len(v):
        print("warning: u and v are of different lengths.")
    if len(u)!=len(z):
        print("warning: u and z are of different lengths.")
    res = res + list(zip(u,v,z))
    print(i,len(u),len(v),len(z))

len(res)
```

```python
meta_data_df = pd.DataFrame(res, columns=['title','relative_link','date'])
meta_data_df.head()
```

```python
## download articles' contents from the harvested links above

filenames = []
for _,link,__ in res:
    url = BASE_URL+link
    html = requests.get(url).content
    sel = Selector(text = html)
    u = sel.xpath(URL2_xpath_content).extract()
    v = ' '.join([x.strip() for x in u]).strip()
    filename = '../data/' + 'www.assarih.com__' + link + '.html'
    filenames.append(filename)
    print(filename)
    with open(filename, 'w', encoding='utf8') as f:
        f.write(v)
```

```python
meta_data_df['filename'] = pd.Series([x[8:] for x in filenames])
```
